```python
import csv
# Generate blank temporary dictionary and a list for all the publication names.
nist_cyber_dict = {}
nist_pub_names_list = []

# Extract data from the NIST-Cybersecurity-Publications.csv file and populate blank dictionary with the data.
with open("NIST-Cybersecurity-Publications.csv", encoding='utf8') as nist_publications_csv:
    nist_pub_details = csv.DictReader(nist_publications_csv)
    
    for row in nist_pub_details:
        alt_name = str(row['Series']) + '-' + str(row['Release Date'])
        name = row['PubID'] if row['PubID'] else (alt_name)
        title = row['Title'] if row['Title'] else 'Not specified'
        keywords = row['Keywords'] if row['Keywords'] else 'Not specified'
        topics = row['Topics'] if row['Topics'] else 'Not specified'
        family = row['Family'] if row['Family'] else 'Not specified'
        url = row['URL'] if row['URL'] else 'Not specified'
        currenturl = row['CurrentURL'] if row['CurrentURL'] else 'Not specified'

        nist_pub = {"PubID": name , "Title": title, "Keywords": keywords, "Topics": topics, "Family": family, "URL": url, "CurrentURL": currenturl}
        nist_cyber_dict[name] = nist_pub
        nist_pub_names_list.append(name)

# Print a list of all the keys within the meta-dictionary []
for key in nist_pub:
    print(key)
# print(nist_pub_names_list)
```

```python
# Breakdown the associated keywords
# Keyword list micro is a dictionary of all nist cyber publications and the associated keywords against them.
keyword_list_dict = {}
# Keyword list macro is a list of all the unique keywords used across the nist cyber publications.
keyword_list_macro = []

# Creating a loop that takes each nist publication name and associated keywords as a key:value pairs.
for i in range(0, len(nist_pub_names_list)):
    name = nist_pub_names_list[i]
    keywords = nist_cyber_dict[name]['Keywords'].split('; ')
    for word in keywords:
        if word not in keyword_list_macro:
            keyword_list_macro.append(word)
    keyword_list_dict[name] = keywords

# Sort the keywords list.
keyword_list_macro.sort()

# Print the length of the keywords list.
print(len(keyword_list_macro))
print(len(keyword_list_dict))
print(keyword_list_dict['NIST SP 800-115'])
```

```python
# Keyword finder function:
## This is a function to be used for searching the nist cyber publication dictionary for occurances of keywords.
def keyword_finder(keyword,dictionary):
    keyword_results = []
    names = nist_pub_names_list
    
    # The loop takes the keyword and searches over all the names in the nist_cyber_dictionary. It returns a list of lists, holding the (name, title, and current link), for every nist publication associated to that keyword.
    for name in names:
        if keyword in dictionary[name]:
            element = [name, nist_cyber_dict[name]['Title'], nist_cyber_dict[name]['CurrentURL']]
            keyword_results.append(element)
    return keyword_results

# Function test for keyword 'security assessment':
keyword_finder('security assessment', keyword_list_dict)
```

```python
# Breakdown the assoicated topics
# Topic list micro is a dictionary of all nist cyber publications and the associated keywords against them
topic_list_dict = {}
# Topic list macro is a list of all the unique keywords used across the nist cyber publications
topic_list_macro = {}
# Sublists for holding the collated topics, and subtopics
collated_topics = []
collated_subtopics = []
# Loop zooming into each nist cyber publication to generate a dictionary of topics per name
for i in range(0, len(nist_pub_names_list)):
    name = nist_pub_names_list[i]
    # Breaking out topics into topic groups and subtopics, first by semi-colon and second by 'colon'
    topics_grouped = nist_cyber_dict[name]['Topics'].split('; ')
#    topics_subtopics = topics_grouped.split(': ')
    # Assigning the topic headers to be the first element of the list from topics_subtopics
    temp_topic_dict_complete = {}
    for group in topics_grouped:
        topics_subtopics = group.split(': ')
        if len(topics_subtopics) == 2:
            topic = topics_subtopics[0]
            subtopics = topics_subtopics[1]
        else:
            topic = topics_subtopics[0]
            subtopics = 'Not specified'
        if topic not in collated_topics:
            collated_topics.append(topic)
        subtopics_split = subtopics.split(', ')
        for element in subtopics_split:
            if element not in collated_subtopics:
                collated_subtopics.append(element)
        
        temp_topic_dict = {}
        temp_topic_dict.setdefault(topic, subtopics_split)
        temp_topic_dict_complete.update(temp_topic_dict)
    
    topic_list_dict[name] = temp_topic_dict_complete 

print(collated_topics)
print(topic_list_dict)
```

```python
# Topic finder function:
## This is a function to be used for searching the nist cyber publication dictionary for occurances of topics.
## The dictionary is semi-fixed to the 'topic_list_dict. The nist_pub_names_list and topic_dictionary are required.
def subtopic_finder(subtopic, dictionary):
    subtopic_results = []
    names = nist_pub_names_list
    for key in topic_dictionary:
        if subtopic in topic_dictionary[key]:
            topic = key
    # The loop takes the topic(subtopics) and searches over all the names in the nist_cyber_dictionary. It returns a list of lists, holding the (name, title, and current link), for every nist publication associated to that subtopic.
    for name in names:
        for topic in dictionary[name]:
            if subtopic in dictionary[name][topic]:
                element = [name, nist_cyber_dict[name]['Title'], nist_cyber_dict[name]['CurrentURL']]
                subtopic_results.append(element)
            else:
                continue
    return subtopic_results

#  A test run through of the subtopic finder.
print(subtopic_finder('big data', topic_list_dict))
```

```python
# Breakdown the associated family
# family list micro is a dictionary of all nist cyber publications and the associated family against them.
family_list_dict = {}
# family list macro is a list of all the unique families used across the nist cyber publications.
family_list_macro = []

# Creating a loop that takes each nist publication name and associated family as a key:value pairs.
for i in range(0, len(nist_pub_names_list)):
    name = nist_pub_names_list[i]
    families = nist_cyber_dict[name]['Family'].split('; ')
    for family in families:
        if family not in family_list_macro:
            family_list_macro.append(family)
    family_list_dict[name] = families

family_list_macro.sort()

print(len(family_list_macro))
print(len(family_list_dict))
print(family_list_dict['NIST SP 800-115'])
```
